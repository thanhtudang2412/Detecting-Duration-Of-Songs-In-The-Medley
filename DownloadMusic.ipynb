```

import openpyxl

from __future__ import unicode_literals
import youtube_dl

archives = "/content/drive/MyDrive/Project/LTM/NHAC_TINH_TUYEN_CHON/"

fileName = archives + "ListenList.xlsx"
begin = 505
end = 529
lane = [30,32,33,36,37,38,41,43,45,46,47,50,51,52,53,54,56,60,84,85,86]
try:
  book = openpyxl.load_workbook(fileName)
except:
  book = openpyxl.Workbook()

sheet = book.active
#lane = range(begin,end)

for walk in lane:
    row = walk + 1
    link = sheet.cell(row=row, column=6).value
    ydl_opts = {
        'format': 'bestaudio/best',
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '64',
        }],
        'outtmpl': '/content/drive/MyDrive/Project/LTM/NHAC_TINH_TUYEN_CHON' + '/'+str(int(sheet.cell(row=row, column=1).value))+'.%(ext)s'
    }

    with youtube_dl.YoutubeDL(ydl_opts) as ydl:
        ydl.download([link])
    sheet.cell(row=row, column=7).value = 'ok'
    book.save(fileName)
print("Done.")
```
