```
from pydub import AudioSegment,silence
import openpyxl
import datetime
from mutagen.mp3 import MP3

archives = "/content/drive/MyDrive/Project/LTM/NHAC_TINH_TUYEN_CHON/"
fileName = archives + "Listened.xlsx"
begin = 88
end = 101
thresh = 24
try:
  book = openpyxl.load_workbook(fileName)
except:
  book = openpyxl.Workbook()
sheet = book.active
row = sheet.max_row + 1 
column = 1
#lane = range(begin,end)
lane = [60,84,85,86]

for walk in lane:
  myaudio = intro = AudioSegment.from_mp3(archives + str(int(walk)) + ".mp3")
  audio = MP3(archives + str(int(walk)) + ".mp3")

  print("Pass phase 1/" + str(int(walk)) + "/" + str(int(end)))
  dBFS=myaudio.dBFS
  print("Pass phase 2/" + str(int(walk)) + "/" + str(int(end)))
  silencemark = silence.detect_silence(myaudio, min_silence_len=3000, silence_thresh=dBFS-thresh)
  print("Pass phase 3/" + str(int(walk)) + "/" + str(int(end)))
  pre = "0:00:00"
  run = walk + 0.01
  for stop,start in silencemark:
    start = start/1000
    sheet.cell(row=row, column=column).value = walk
    sheet.cell(row=row, column=column + 1).value = run
    sheet.cell(row=row, column=column + 2).value = pre + " - " + str(datetime.timedelta(seconds=int(start)))
    pre = str(datetime.timedelta(seconds=int(start)))
    row += 1
    run += 0.01
  if pre != str(datetime.timedelta(seconds=int(audio.info.length))):
    sheet.cell(row=row, column=column).value = walk
    sheet.cell(row=row, column=column + 1).value = run
    sheet.cell(row=row, column=column + 2).value = pre + " - " + str(datetime.timedelta(seconds=int(audio.info.length)))
    row += 1
  book.save(fileName)

print("Done.")
```
